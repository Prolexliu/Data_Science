### Methods: Write up the methods used in your analysis

Answer:

We will be using Logistic Regression with dummies variables to analysis how the variables influence the probability of admission. The package we will be using is statsmodels.

First we will need to create dummies variable for "prestige". We only want logistic regression to only consider binary option (Ture/False) instead of comparing the value, for example, prestige 4 have a higher value then prestige 3. 

Second, We're going to add a constant term for our Logistic Regression. The statsmodels function we're going to be using requires that intercepts/constants are specified explicitly

Finally, the probablility will be exp() of the coeffiecent from the models, and we will use this model the predict other dataset
